# Generating a mini-corpus

## Imports

```python
import json
import twitter
import urllib.parse
```

## Connect to Twitter API

```python
# Load tokens from file
with open('../data/tokens.json', 'r') as f:
    tokens = json.load(f)
```

```python
# Connect to API
api = twitter.Api(
    consumer_key=tokens['consumer_key'],
    consumer_secret=tokens['consumer_secret'],
    access_token_key=tokens['access_token_key'],
    access_token_secret=tokens['access_token_secret'],
    tweet_mode='extended'
)
```

## Querying the API

Query:
- Terms (OR):
    - #Covid19
    - #Covid-19
    - #C19
    - #coronavirus 
    - #covid2019
    - #flattenthecurve
- lang: EN
- since: 2020-01-01
- count: ?

```python
results = api.GetSearch(
    raw_query=urllib.parse.urlencode({
        'q':'covid OR cornavirus OR c19 OR flattenthecur  -filter:retweets',
        'lang':'en',
        'since':'2020-01-01',
        'result_type': 'recent',
        'count': 100
    })
)
```

```python
for entry in results:
    print(dir(entry))
    break
```
